```python
x=data.drop('benign_0__mal_1',axis=1).values
y=data['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaled=MinMaxScaler()
```

```python
x_train=scaled.fit_transform(x_train)
x_test=scaled.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_test.shape
```

```python
model=Sequential()
model.add(Dense(30,activation='relu'))

model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping 
```

```python
early_stop= EarlyStopping(monitor='val_loss',patience=25,mode='min')
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),callbacks=[early_stop],epochs=450)
```

```python
losses=pd.DataFrame(model.history.history)
losses.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model=Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),callbacks=[early_stop],epochs=450)
```

```python
losses=pd.DataFrame(model.history.history)
losses.plot()
```

```python
predict=model.predict_classes(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,predict))
```
